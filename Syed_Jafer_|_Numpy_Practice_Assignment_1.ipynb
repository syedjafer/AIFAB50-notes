```python
# 7) create 4X4 matrix using random numbers within the range of 10 to 50 and store it in variable arr
# 8) 2X2 Maxpooling operation on arr, 
# 9) 2X2 Average pooling on arr
```

```python
# 7) create 4X4 matrix using random numbers within the range of 10 to 50 and store it in variable arr
mat_4_4 = np.random.randint(low=10, high=51, size=(4,4))
print(mat_4_4)
```

```python
mat_4_4.flatten()
```

```python
# 8) 2X2 Maxpooling operation on arr, 

temp = []
flt_mat_4_4 = mat_4_4.flatten()
total_length = len(flt_mat_4_4)
for itr, num in enumerate(flt_mat_4_4):
    if itr+5 > total_length-1:
        break
    _max = np.max([flt_mat_4_4[itr], flt_mat_4_4[itr+1], flt_mat_4_4[itr+4], flt_mat_4_4[itr+5]])
    temp.append(_max)
res = np.array(temp)
res.resize((3,3))
print(res)


    
    

    
```

```python
# 9) 2X2 Average pooling on arr

temp = []
flt_mat_4_4 = mat_4_4.flatten()
total_length = len(flt_mat_4_4)
for itr, num in enumerate(flt_mat_4_4):
    if itr+5 > total_length-1:
        break
    _max = np.average([flt_mat_4_4[itr], flt_mat_4_4[itr+1], flt_mat_4_4[itr+4], flt_mat_4_4[itr+5]])
    temp.append(_max)
res = np.array(temp)
res.resize((3,3))
print(res)

```

```python
# R+G
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
c = 0 # Red
tmp_im = np.zeros(img.shape, dtype="uint8")
tmp_im[:,:,c] = img[:,:,c]
tmp_im[:,:,c+1] = img[:,:,c+1]
ax.imshow(tmp_im)
ax.set_axis_off()
```

```python
# R + B
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
tmp_im = np.zeros(img.shape, dtype="uint8")
tmp_im[:,:,0] = img[:,:,0]
tmp_im[:,:,2] = img[:,:,2]
ax.imshow(tmp_im)
ax.set_axis_off()
```
